## task 1: to find the derivative for given equation 2x^2+3y

```python
import sympy as sp
x,y=sp.symbols('x,y')
```

```python
f=2*x**2+3*y  ##equation
##derivation formula apply
df_dx=sp.diff(f,x)
df_dy=sp.diff(f,y)
print(f"The equation is:{f}")
print("The derivation of x is df/dx is=",df_dx)
print("The derivation of y is df/dy is=",df_dy)

```
