## Task2: To find the derivation of given formula 4x^2+2y^3x

```python
import sympy as sp
x,y=sp.symbols('x,y')
f=4*x**2*y+2*y**3*x  ##equation
## derivation calculation
df_dx=sp.diff(f,x)
df_dy=sp.diff(f,y)
print("The equation is:",f)
print("The derivation of x is df/dx:",df_dx)
print("The derivation of y is df/dy:",df_dy)
```
